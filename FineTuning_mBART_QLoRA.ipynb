```python
from datasets import load_dataset

ds = load_dataset("Ketan3101/English-Hindi-Translation")
dataset = ds["train"].shuffle(seed=42).select(range(3500))

dataset

```

```python
dataset[:5]
```

```python
data = dataset.train_test_split(test_size=0.3, seed=42)

```

```python
data
```

```python
from transformers import AutoTokenizer

model_name = "facebook/mbart-large-50-many-to-many-mmt"
tokenizer = AutoTokenizer.from_pretrained(model_name)

tokenizer.src_lang = "en_XX"
tokenizer.tgt_lang = "hi_IN"

```

```python
Example_text = "The accused man has been identified as Sarvesh Kumar."
tokenizer(Example_text, padding="max_length", truncation=True, max_length=64)

```

```python
def tokenize_function(examples):
    inputs = tokenizer(examples["english"], padding="max_length", truncation=True, max_length=64)
    labels = tokenizer(examples["hindi"], padding="max_length", truncation=True, max_length=64)
    inputs["labels"] = labels["input_ids"]
    return inputs

# Apply tokenization
dataset = data.map(tokenize_function, batched=True)

```

```python
dataset
```

```python
from transformers import AutoModelForSeq2SeqLM, BitsAndBytesConfig
from peft import LoraConfig, get_peft_model, TaskType
import torch

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True
)

model = AutoModelForSeq2SeqLM.from_pretrained(model_name, quantization_config = bnb_config)

peft_config = LoraConfig(
    r=16,
    lora_alpha=32,
    lora_dropout=0.1,
    bias="none",
    task_type=TaskType.SEQ_2_SEQ_LM,
    target_modules=["self_attn.q_proj", "self_attn.v_proj"]
)

model = get_peft_model(model, peft_config)

```

```python
from transformers import DataCollatorForSeq2Seq
data_collator = DataCollatorForSeq2Seq(tokenizer, model = model)

```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./translation_model_qLORA",
    evaluation_strategy="epoch",
    logging_strategy="epoch",
    report_to="none",
    learning_rate=1e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=5,
    weight_decay=0.01,
    save_total_limit=1,
    logging_dir="./logs"
)

```

```python
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset["train"],
    eval_dataset=dataset["test"],
    tokenizer=tokenizer,
    data_collator=data_collator
)

```

```python
trainer.train()
```

```python
results = trainer.evaluate()
results

```

```python
def translate_text(text, model, tokenizer):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=64)
    inputs = {key: val.to(device) for key, val in inputs.items()}

    # ✅ Force the model to output Hindi
    forced_bos_token_id = tokenizer.lang_code_to_id["hi_IN"]

    output = model.generate(**inputs, max_length=64, forced_bos_token_id=forced_bos_token_id)
    translated_text = tokenizer.decode(output[0], skip_special_tokens=True)
    return translated_text


# ✅ Test the model with 3 sample sentences
test_sentences = [
    "Hello, how are you?",
    "My name is Jinil Patel.",
    "I am a student, and I want to join as an AI engineer in your company."
]

for sentence in test_sentences:
    print(f"English: {sentence}")
    print(f"Hindi: {translate_text(sentence, model, tokenizer)}\n")
```

```python
import torch
from peft import PeftModel
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

# ✅ Set Base Model & Adapter Path
BASE_MODEL = "facebook/mbart-large-50-many-to-many-mmt"  # Keep as per your model
ADAPTER_PATH = "/content/translation_model_qLORA/checkpoint-1535"  # Your LoRA adapter path
MERGED_MODEL_PATH = "/content/translation_model_merged"

# ✅ Load Base Model
print("Loading base model...")
base_model = AutoModelForSeq2SeqLM.from_pretrained(BASE_MODEL)
tokenizer = AutoTokenizer.from_pretrained(BASE_MODEL)

# ✅ Load LoRA Adapter
print("Loading LoRA adapter...")
model = PeftModel.from_pretrained(base_model, ADAPTER_PATH)

# ✅ Merge LoRA Adapter into Base Model
print("Merging LoRA adapter...")
model = model.merge_and_unload()

# ✅ Save the Merged Model
print("Saving merged model...")
model.save_pretrained(MERGED_MODEL_PATH)
tokenizer.save_pretrained(MERGED_MODEL_PATH)

print(f"✅ LoRA adapters merged and model saved successfully at {MERGED_MODEL_PATH}")
```
